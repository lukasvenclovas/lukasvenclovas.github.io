```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ei = pd.read_csv('ei_2021_12_31.csv')
```

```python
print(ei.columns)
```

```python
ei.describe()
```

```python
ei.head(10)
```

```python
grupavimas = ei.groupby('rusis')['iskaitinis'].sum().reset_index()

plt.figure(figsize = (30,6))
plt.bar(grupavimas['rusis'], grupavimas['iskaitinis'], color = 'purple')
plt.title('Iskaitiniu ivykiu skaicius pagal eismo ivykiu rusis')
plt.xlabel('Eismo ivykiu rusis')
plt.ylabel('Iskaitiniu ivykiu skaicius')
plt.show()
```

```python
#Ivykiu kiekis pagal menesi
ei['registravimoData'] = pd.to_datetime(ei['registravimoData'])

ei['Menuo'] = ei['registravimoData'].dt.month

sk_per_menesi = ei.groupby('Menuo').size()

plt.plot(sk_per_menesi.index,sk_per_menesi.values, color = 'orange', marker = 'o')
plt.xlabel('Menuo')
plt.ylabel('Ivykiu kiekis')
plt.title('Ivykiu kiekis pagal menesi')
plt.grid(True)
plt.show()

```

```python
#Ivykiu kiekis pagla eismo ivykiu rusis

rusis = ei.groupby('rusis').size()

rusis.plot(kind = 'bar', linestyle = 'dotted', color = 'brown')
plt.title('Ivykiu kiekis pagal eismo ivykiu rusis')
plt.xlabel('Eismo ivykiu rusys')

plt.ylabel('Ivykiai (kiekis)')
plt.show()

```

```python
#Dazniausiai pasikartojancios eismo ivykiu rusis 

dazn_rusis = ei['rusis'].value_counts()

dazn_rusis.plot(kind = 'bar', color = 'skyblue')
plt.title('Dazniausios eismo ivykiu rusys')
plt.xlabel('Eismo ivykiu rusis')
plt.ylabel('Pasikartojimu skaicius')
plt.show()
```

```python
#Zuvusiu asmenu kiekis pagal menesius


zuvusiu_asm_sk = ei.groupby('Menuo')['zuvusiuSkaicius'].sum()
suzeistu_asm_sk = ei.groupby('Menuo')['suzeistuSkaicius'].sum()

plt.plot(zuvusiu_asm_sk.index,zuvusiu_asm_sk.values, color = 'blue', marker = 'o', linestyle = 'dotted')
plt.title('Zuvusiu asmenu kiekis pagal menesius')
plt.xlabel('Menuo')
plt.ylabel('Asmenu skaicius')
plt.grid(True)
plt.show()
```

```python
#Suzeistu asmenu kiek pagal menesius

plt.plot(suzeistu_asm_sk.index,suzeistu_asm_sk.values, color = 'green', marker = 'o', linestyle = 'dashdot')
plt.title('Suzeistu asmenu kiekis pagal menesius')
plt.xlabel('Menuo')
plt.ylabel('Asmenu skaicius')
plt.grid(True)
plt.show()
```

```python
#zuvusiu vaiku kiekis per menesius

zuve_vaikai = ei.groupby('Menuo')['zuvVaiku'].sum()

plt.figure(figsize = (10,6))
plt.bar(zuve_vaikai.index,zuve_vaikai.values, color = 'Yellow')
plt.title('Zuvusiu vaiku skaicius per menesius')
plt.xlabel('Menuo')
plt.ylabel('Zuvusiu vaikus skaicius')

plt.show()
```

```python
#Suzeistu vaiku kieki per menesius

suzeisti_vaikai= ei.groupby('Menuo')['suzeistaVaiku'].sum()

plt.figure(figsize = (10,6))
plt.bar(suzeisti_vaikai.index,suzeisti_vaikai.values, color = 'grey')
plt.title('Suzeistu vaiku skaicius per menesius')
plt.xlabel('Menuo')
plt.ylabel('Suzeistu vaikus skaicius')
plt.show()
```
